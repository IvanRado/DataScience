# Cleaning the Data

Before we are able to perform any real data analysis, we will need to clean our data. To that end, we need to eliminate incorrect or useless data. For us, this is data that falls into two categories:
    
  1. Paid apps. Remember, we are looking at apps for in-app advertising, paid apps will not be having such advertisements
  2. Apps for non English speakers - our target market is English speakers and thus we are not concerned with 


```python
# The shitty row is 10472
print(gp_data[10473])
del(gp_data[10473])
```

The Google Play Market has duplicate app data. We can see this by sorting the apps by uniqueness in a dictionary

```python
duplicate_apps_gp = []
unique_apps_gp = []

for row in gp_data[1:]:
    name = row[0]
    if name in unique_apps_gp:
        duplicate_apps_gp.append(name)
    else:
        unique_apps_gp.append(name)

duplicate_apps_apple = []
unique_apps_apple = []

for row in apple_data[1:]:
    name = row[1]
    if name in unique_apps_apple:
        duplicate_apps_apple.append(name)
    else:
        unique_apps_apple.append(name)

        

print('Number of duplicate apps in Google Play Market:' + str(len(duplicate_apps_gp)))
print('\n')
print('Number of duplicate apps in Apple Store:' + str(len(duplicate_apps_apple)))
```

We want to remove duplicate entries. There are a few different ways that we could do this. One option is to keep the entry with the most reviews, as this is guaranteed to be the most recent version of the app and thus most accurately reflects the current state of affairs.

```python
reviews_max_gp = {}

for row in gp_data[1:]:
    name = row[0]
    n_reviews = float(row[3])
    
    if name in reviews_max_gp and reviews_max_gp[name] < n_reviews:
        reviews_max_gp[name] = n_reviews
    elif name not in reviews_max_gp:
        reviews_max_gp[name] = n_reviews

print(len(reviews_max_gp))

android_clean = []
already_added_gp = []

for row in gp_data[1:]:
    name = row[0]
    n_reviews = float(row[3])
    
    if n_reviews == reviews_max_gp[name] and name not in already_added_gp:
        android_clean.append(row)
        already_added_gp.append(name)
        
print(android_clean[0:4])

print(len(android_clean))


reviews_max_apple = {}

for row in apple_data[1:]:
    name = row[1]
    n_reviews = float(row[5])
    
    if name in reviews_max_apple and reviews_max_apple[name] < n_reviews:
        reviews_max_apple[name] = n_reviews
    elif name not in reviews_max_apple:
        reviews_max_apple[name] = n_reviews

print(len(reviews_max_apple))

apple_clean = []
already_added_apple = []

for row in apple_data[1:]:
    name = row[1]
    n_reviews = float(row[5])
    
    if n_reviews == reviews_max_apple[name] and name not in already_added_apple:
        apple_clean.append(row)
        already_added_apple.append(name)

print(len(apple_clean))
```

Next we want to deal with apps that are not usable by English speakers. One way that we can identify such apps is by checking the numerical value of the characters. All English letters and traditional symbols can be represented bt ASCII encoding where the highest character figure is 127. Thus, if a title has a character whose value is above this threshold, we can surmise that it does not belong to the English alphabet and is most likely not usable by an English speaking user.

```python
def englishSpeaking(string):
    char_count = 0
    for letter in string:
        if ord(letter) > 127:
            char_count += 1
            
    if char_count > 3:
        return False
    else:
        return True

print(englishSpeaking('Instagram'))
print(englishSpeaking('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(englishSpeaking('Docs To Go™ Free Office Suite'))
print(englishSpeaking('Instachat 😜'))
```

We now apply the above function to both datasets

```python
android_english = []
apple_english = []

for row in android_clean:
    name = row[0]
    
    if englishSpeaking(name) == True:
        android_english.append(row)

for row in apple_clean:
    name = row[1]
    
    if englishSpeaking(name) == True:
        apple_english.append(row)
        
print(len(android_english))
print(len(apple_english))
      
```

Next we want to make sure that we isolate all the free apps, since we plan to include in-app advertising.

```python
android_clean_free = []
apple_clean_free = []

for row in android_english:
    price = row[7]
    
    if price == '0':
        android_clean_free.append(row)
    
    
for row in apple_english:
    price = float(row[4])
    
    if price == 0.0:
        apple_clean_free.append(row)
    

print(len(android_clean_free))
print(len(apple_clean_free))
```

# Most Common Apps by Genre

To minimize risks and overhead, our validation strategy for an app idea is comprised of three steps:

1. Build a minimal Android version of the app, and add it to Google Play.
2. If the app has a good response from users, we then develop it further.
3. If the app is profitable after six months, we also build an iOS version of the app and add it to the App Store.

Because our end goal is to add the app on both the App Store and Google Play, we need to find app profiles that are successful on both markets. For instance, a profile that might work well for both markets might be a productivity app that makes use of gamification.

Let's begin the analysis by getting a sense of what are the most common genres for each market. For this, we'll need to build frequency tables for a few columns in our data sets.

```python
gp_genres = {}
gp_genres_list = []

for row in android_clean_free:
    genre = row[9]
    
    if genre in gp_genres:
        gp_genres[genre] += 1
    else: 
        gp_genres[genre] = 1
    
    gp_genres_list.append(genre)
    
# print(gp_genres)

apple_genres = {}
apple_genres_list = []

for row in apple_clean_free:
    genre = row[11]
    
    apple_genres_list.append(genre)
    
    if genre in apple_genres:
        apple_genres[genre] += 1
    else:
        apple_genres[genre] = 1

# print(apple_genres_list[0:4])
```

```python
def freq_table(dataset, index):
    table = {}
    total = len(dataset)
    for row in dataset:
        entry = row[index]
        
        if entry in table:
            table[entry] += 1
        else: 
            table[entry] = 1
            
    for key in table:
        table[key] /= total
        table[key] *= 100
        
    return table

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
   
print('Android breakdown: \n')
display_table(android_clean_free, 1)
print('\n')
display_table(android_clean_free, 9)
print('\n')
print('Apple Breakdown: \n')
display_table(apple_clean_free, 11)
        
```

Analysing the two datasets we see two very different landscapes. In the Apple Store, a majority of the free apps are actually games (~58.14%), followed far behind by entertainment at 7.89%. Apps focused on more practical matters, such as education, productivity or health and fitness seem to take up a much smaller slice of the pie. The Google Play Store has a much different distribution, with the Family genre dominating, but at only 18.9%. Further investigation reveals that about ~60% of apps here are in fact games for kids, but this nonetheless leaves games at about 20% of the app market as opposed to the 58.14% of the Apple Store.

Overall the marketplace on the Apple Store seems very geared towards games and entertainment while the Google Play marketplace seems like it has a much greater balance between entertainment and practical apps. Looking at the Genres column, we see a much more granular breakdown of the data; while the relation between this and the Category column is not precisely defined we will be working with the Genre column for a more big picture overview.

To make a good recommendation, we must dig a bit deeper into the data. While the number of entertainment apps may be high in both cases, this does not necessarily mean that each individual app is more popular. 


# Most Popular Apps

### Apple Store

One way we can look at which apps are the most popular is by considering how many times an app has been installed. This data is readily available in the Google Play data set but is missing from the Apple Store one. To work around this, we can use the number of total reviews that an app has instead, as we can safely assume that an app has been downloaded at least once by a user if it has been reviewed by them.

```python
print('Average number of reviews by genre: \n')
for genre in apple_genres:
    
    total = 0
    len_genre = 0
    
    for row in apple_clean_free:
        genre_app = row[11]
        
        if genre_app == genre:
            ratings = float(row[5])
            total += ratings
            len_genre += 1
            
    avg_ratings = total/len_genre
    print(genre, avg_ratings)
```

On average, navigation apps have the highest number of user reviews. These may be skewed by Google Maps and Waze which combine for close to half a million reviews.

```python
for app in apple_clean_free:
    if app[-5] == 'Navigation':
        print(app[1], ':', app[5])
```

We can notice this same problem when it comes to social networking and music apps; the average is heavily influenced by giants such as Facebook, Instagram, Spotify and Pandora. 

The reference category suffers a similar issue, though here the dominant apps are the Bible and Dictionary.com. Seeing this and that books have a sizable amount of average reviews we might posit that opportunity exists here. Other popular categories, such as weather, food, and finance generally do not require a lot of in-app time or rely on a separate service which is not in the scope of our project.

Thus, we may suggest that an app for individual or book collections may be viable option. Such apps would see the user having the app open for longer periods of time so as to generate more ad time, and would not have to struggle to breakthrough a market as oversaturated as the gaming market on the Apple Store. 

Now let's take a look at Google Play apps.

### Google Play

As mentioned above, we actually have install information for Google Play apps. However, this information is discrete, being lumped into categories such as 100+, 10,000+, 100,000+ etc. Since we are just doing a high level overview, this level of precision is enough for us. However, we will reformat these strings into floats so that we may do some computations on the average number of installs. To do so we will remove the commas and plus signs and then cast the values as floats. The results can be seen below.

```python
gp_categories = {}
gp_categories_list = []

for row in android_clean_free:
    genre = row[1]
    
    if genre in gp_categories:
        gp_categories[genre] += 1
    else: 
        gp_categories[genre] = 1
    
    gp_categories_list.append(genre)
    
print('Average number of installs by category: \n')
for category in gp_categories:
    total = 0
    len_category = 0
    
    for row in android_clean_free:
        category_app = row[1]
        
        if category_app == category:
            installs = row[5]
            installs = installs.replace('+','')
            installs = installs.replace(',','')
            installs = float(installs)
            total += installs
            len_category += 1
            
    avg_installs = total/len_category
    print(category, avg_installs)
                                        
```

Here we see a similar conundrum to the Apple Store. While communication apps have on average the highest number of installs, they are completely dominated by giants such as Skyle, Facebook Messenger, and Gmail that we have no chance at competing with.

```python
for app in android_clean_free:
    if app[1] == "COMMUNICATION" and (app[5] == '1,000,000,000+'
                                     or app[5] == '500,000,000+'
                                     or app[5] == '100,000,000+'):
        print(app[0], ":", app[5])    
```

We can reasonably assume that removing these giants (with whom we can't compete) would significantly bring down the average. A similar pattern can be seen in the video players, social, and productivity apps which are dominated by the likes of YouTube, Facebook, and Dropbox.

Since we are looking at a profitable app profile for both Google Play and the Apple Store, we can further investigate the Books and References genre since it previously showed some potential

There are a variety of apps included here, including apps for processing and reading ebooks, audiobooks, libraries, dictionaries, etc. While there are still some extremely popular apps that skew the average, they less numerous and less domineering than in other genres. Apps that are popular but do not dominate in this space are generally dedicated to dealing with ebooks or are collections of libraries or dictionaries.
